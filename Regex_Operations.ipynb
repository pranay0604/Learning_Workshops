```python
import re
```

**re.match**

- Matches a pattern in a given string

- at start of a string

```python
pattern = r"hello"
str1 = "hello world"

pattern in str1
# 'hello' in "hello world"
```

```python
pattern = r"hello"
str1 = "hello world"

res = re.match(pattern,str1)
res.group()
```

```python
pattern = r'hi'
str1 = "hi how are you"
re.match(pattern,str1)
```

```python
pattern = r'hi'
str1 = "hi how are you"
re.match(pattern,str1).group()
```

```python
pattern = r'you'
str1 = "hi how are you"
re.match(pattern,str1)
# no output
```

```python
pattern = r'hello'
str1 = "world hello"
res = re.match(pattern,str1)
res # no output
```

**re.search**

- Searches pattern for anywhere in a given string

```python
pattern = r'hello'
str1 = "world hello"
res = re.search(pattern,str1)
res 
```

```python
pattern = r'hello'
str1 = "world hello"
res = re.search(pattern,str1)
res.group()
```

```python
pattern = r'hi'
str1 = "hi how are you"
re.search(pattern,str1)
```

```python
pattern = r'hi'
str1 = "hi how are you"
re.search(pattern,str1).group()
```

**re.findall**

- using a \d+ pattern

- \d indicates digit match between 0 to 9

- plus indicates one or more occurences of a digit

```python
str1 = "the price is 100 and the tax is 20"

ans = []
str1=str1.split()

for i in str1:
    if i.isdigit():
        ans.append(i)
ans
```

```python
str1 = "the price is 100 and the tax is 20"
pattern = r"\d+"

res = re.findall(pattern,str1)
res
```

```python
str1 = "order number is 1234 shipped on 15th Feb 2026"
pattern = r"\d+"

res = re.findall(pattern,str1)
res
```

```python
str1 = "order number is 1234 shipped on 15th Feb 2026"

ans = []
str1=str1.split()

for i in str1:
    if i.isdigit():
        ans.append(i)
ans
```

**Email Valid Use Case**

- digit

- letters

- special characters

```python
# ^ : used for start of the string
# $ : used for end of the string

# [a-zA-Z0-9_.+-] : pattern if any letter in email id have this not only one character, 
#                   many characters for that we need to add a + sign at last
# [a-zA-Z0-9_.+-]+
# till now the pattern is ^[a-zA-Z0-9_.-+]+

# @ : email id has a symbol
# till now the pattern is ^[a-zA-Z0-9_.-+]+@

# [a-zA-Z0-9-]+
# till now the pattern is ^[a-zA-Z0-9_.-+]+@[a-zA-Z0-9-]+

# \. 
# till now the pattern is ^[a-zA-Z0-9_.-+]+@[a-zA-Z0-9-]+\.

# com
# # till now the pattern is ^[a-zA-Z0-9_.-+]+@[a-zA-Z0-9-]+\.[a-zA-Z]{2,}

# $ : end of the string

# final pattern is 

pattern = r"^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z]{2,}$"
```

```python
pattern = r"^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z]{2,}$"
email = 'pranay.sontakke@gmail.com'
bool(re.match(pattern,email))
```
