## Regex Operations


### Matching a Sequence of 10 Numbers

**d and D family**

- Here we are focusing on identifying the digits

    - \d : 1 digit
    
    - \d+ : More than 1 digit
    
    - \d{3} : Exactly 3 digits
 
        - Ex: 1234 : 123 will come only 
    
    - \d{3,} : Atleast 3 digits
 
    - \d{2,4} : Between 2 to 4 digits

```python
import re

pattern = r'\d'
str1 = 'my age is 50'
re.findall(pattern,str1)
```

```python
import re

pattern = r'\d+'
str1 = 'my age is 50, mobile no is 1234567'
re.findall(pattern,str1)
```

```python
import re

pattern = r'\d{3}'
str1 = 'my age is 5000'
re.findall(pattern,str1)
```

```python
import re

pattern = r'\d{3}'
str1 = 'My mobile number is 1234567890'
re.findall(pattern,str1)
```

```python
import re

pattern = r'\d{2,4}'
str1 = 'My mobile number is 1234567890'
re.findall(pattern,str1)
```

```python
import re

pattern = r'\d{3,}'
str1 = 'My mobile number is 1234567890'
re.findall(pattern,str1)
```

```python
str1 = 'My mobile number is 123-456-7890'
pattern = r'\d{3}-\d{3}-\d{4}'
re.findall(pattern,str1)
```

```python
str1 = 'My mobile number is (123)-456-7890'
pattern = r'\(\d{3,}\)-\d{3,}-\d{4}'
re.findall(pattern,str1)
```

```python
str1 = 'My mobile number is +91-123-456-7890'
pattern = r'\+\d{2}-\d{3}-\d{3}-\d{4}'
re.findall(pattern,str1)
```

```python
str1 = 'My mobile number is +91-123-456-7890'
pattern = r'\+91-\d{3}-\d{3}-\d{4}'
re.findall(pattern,str1)
```

**w and W family**

- \w : a-z and A-Z and 0-9 and _

    - single character
    
    - ex: str1 = 'ram' : r,a,m

- \w+ : more than one character

    - ex: str1 = 'ram' : ram

- \W : single punctuations

- \W+ : multiple punctuations

```python
s1 = "Ramesh_25"
pattern = r'\w'
re.findall(pattern,s1)
```

```python
s2 = "My name is Ramesh"
pattern = r'\w+'
re.findall(pattern,s2)
```

```python
s3 = 'Hello@123'
pattern = r'\W'
re.findall(pattern,s3)
```

```python
s4 = 'Hello#!world@123'
pattern = r'\W+'
re.findall(pattern,s4)
```

**s and S family**

- \s : identify space, tabs, new line

    - single space only

- \s+ : identify multiple spaces, tabs, new lines

- \S : non-space character

    - ex: s1 = 'Hi A' = H i and A

- \S+ : non-space characters word will come

```python
s1 = 'good morning'
pattern = r'\s'
re.findall(pattern,s1)
```

```python
s2 = 'good  morning'
pattern = r'\s+'
re.findall(pattern,s2)
```

```python
s3 = 'Hi A'
pattern = r'\S'
re.findall(pattern,s3)
```

```python
s4 = 'my name is Ram'
pattern = r'\S+'
re.findall(pattern,s4)
```

**Symbols**

- Caret ^ : Start of the string

- Dollar $ : End of the string

```python
s1 = 'my name is Ram'
pattern = r'^my'
re.findall(pattern,s1)
```

```python
s1 = 'my name is Ram'
pattern = r'Ram$'
re.findall(pattern,s1)
```

```python
s1 = 'my name is Ram'
pattern = r'm$'
re.findall(pattern,s1)
```

```python
s1 = '123456789'
pattern = r'^1\d+9$'
re.findall(pattern,s1)
```

```python
s1 = '123456789'
pattern = r'^\d+$'
re.findall(pattern,s1)
```

```python
s1 = 'cat bat mat sat'
pattern = r'.at'
re.findall(pattern,s1)
```

```python
s1 = 'cat bat mat sat fat moon sun'
pattern = r'.at'
re.findall(pattern,s1)
```

```python
s1 = 'Mob no is 123456789'
pattern = r'1\d+9'
re.findall(pattern, s1)
```

```python
s1 = 'cat bat mat sat fat moon sun'
pattern = r'[cbmsf]at'
re.findall(pattern,s1)
```
