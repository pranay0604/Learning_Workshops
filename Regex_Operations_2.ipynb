Data Cleaning
----

```python
import re

text='this  is   a    sample    text'
re.sub(r'\s+'," ",text)
```

```python
text='apple;banana|grape,orange'
pattern='[;|,]'
re.split(pattern,text)
```

```python
text='apple;banana|grape,orange'
text.split(';')
```

HTML Parser
---

```python
html="""
 <a href="https://www.atyeti.com/">Atyeti</a>
 <a href="https://www.openai.com/">Atyeti</a>
 """

pattern='href="(.*?)"'
re.findall(pattern,html)
```

File Validation
---

```python
pattern=r'.*\.(jpg|jpeg|png|csv)$'
bool(re.match(pattern,'image.png'))
```

```python
bool(re.match(pattern,'image.pdf'))
```

```python
bool(re.match(pattern,'image.csv'))
```

```python
bool(re.match(pattern,'image.xlsx'))
```
